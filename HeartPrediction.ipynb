# Heart Disease Prediction

## Importing the Libraries

importing the necessary libraries for building a model

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

## Loading the data

Loading the test data and train data

```python
train = pd.read_csv("train.csv")
test = pd.read_csv('test.csv')
print(train.head())
```

## Working on the train data

```python
print(train.tail())
```

```python
train.info()
```

```python
train.corr()
```

```python
train.describe()
```

```python
x_train = train.drop(['target'], axis=1)
y_train = train['target']

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, train_size=0.7, random_state=12, shuffle=True)
```

```python
x_train.info()
```

## Using Logistic Regression

```python
lrm = LogisticRegression(max_iter=5000)
lrm.fit(x_train, y_train)

print(accuracy_score(y_pred=lrm.predict(x_train), y_true=y_train))
print(accuracy_score(y_pred=lrm.predict(x_test), y_true=y_test))
```

```python
print(lrm)
print('\nFor Train')
print(classification_report(y_pred=lrm.predict(x_train), y_true=y_train))
print('\nFor Test')
print(classification_report(y_pred=lrm.predict(x_test), y_true=y_test))
```

## Improving accuracy of Logistic Regression using GridSearchCV

```python
param_grid = [
    {'penalty' : ['l2'],
     'C' : np.logspace(-4, 4, 20),
     'solver' : ['newton-cg', 'liblinear', 'sag', 'saga'],
     'max_iter' : [500, 1000, 3000, 5000]
     }
]
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(lrm, param_grid= param_grid, cv=3, verbose=True, n_jobs=-1)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_estimator_
```

```python
print(clf.score(x_train, y_train))
print(clf.score(x_test, y_test))
```

## Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

randMod = RandomForestClassifier()
randMod.fit(x_train, y_train)
```

```python
print(accuracy_score(y_pred=randMod.predict(x_test), y_true=y_test))
```

```python
print(classification_report(y_pred=randMod.predict(x_train), y_true=y_train))
```

```python
print(classification_report(y_pred=randMod.predict(x_test), y_true=y_test))
```

## Overcoming overfitting for Random Forest

```python
param_grid = {
    'n_estimators' : [int(x) for x in np.linspace(start=10, stop=80, num=10)],
    'max_features' : ['log2', 'sqrt'],
    'max_depth' : [2, 4],
    'min_samples_split' : [2, 5],
    'min_samples_leaf' : [1, 2],
    'bootstrap' : [True, False]
}
```

```python
n_rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(estimator= n_rf, param_grid= param_grid, cv=3, verbose=2, n_jobs=4)
```

```python
rf_grid.fit(x_train, y_train)
```

```python
rf_grid.best_estimator_
```

```python
print(rf_grid.score(x_train, y_train))
print(rf_grid.score(x_test, y_test))
print(n_rf)
```

## Dumping the model into a pickle file

The model trained is stored in a pickle file `modelFile.pkl` and is used for predicting heart disease of the user.

```python
import pickle
pickle.dump(rf_grid, open('modelFile.pkl', 'wb'))
```

```python
# Sample test using the pickle file

input_data = [72,1,1,180,300,0,1,168,0,1,2,2,3]
input_data = [np.array(input_data)]
features = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
df = pd.DataFrame(input_data, columns=features)

model = pickle.load(open('modelFile.pkl', 'rb'))
res = model.predict(df)
if res==1:
    value = 'high risk'
else:
    value = 'low risk'
```
